```python
caminho = 'wc2.1_10m_tavg_01.csv'
```

```python
def dataframe_pd(path):
  df_pandas = None
  df_pandas = pd.read_csv(caminho)
```

## Retorna o tempo de execução do dataframe criado utilizando o Pandas em segundos no arquivo txt tempos_execucao



```python
with open('tempos_execucao_pandas.txt', 'w') as arquivo:
  for test in range(32):
    tempo_execucao = %timeit -o -r 1 -n 1 -q dataframe_pd(caminho)
    tempo_execucao_str = str(tempo_execucao.best) + '\n'
    arquivo.write(tempo_execucao_str)
```
